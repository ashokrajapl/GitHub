```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_pairs=[]
    corr_matrix = dataset
    for i in range(len(corr_matrix.columns)):
#         print('i',i)
        for j in range(i):
#             print('j',j)
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                pair=[corr_matrix.columns[i],corr_matrix.columns[j],corr_matrix.iloc[i, j]]
                corr_pairs.append(pair)
                col_corr.add(colname)        
    return col_corr,corr_pairs
```

```python
corr_features,corr_pair = correlation(corr_df, 0.9)
corr_features
corr_pair
```

- The list of highly correlated features are mentioned above.Any one featues in the pair features can be removed while training a model but since I am doig dimensional reduction anyways I am keeping these features 

<h3> Z Score </h3>

```python
x=car_data.drop('class',axis=1)
y=car_data['class']
from scipy import stats
x=stats.zscore(x)
```

<h3> Train Test Split </h3>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

<h2>2. Classifier </h2>

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm, datasets
svc = svm.SVC()
parameters = {'kernel':['linear','rbf'],'C':[1, 2,4,6,10,20]}
clf = GridSearchCV(svc,parameters)
clf.fit(x_train,y_train)
results = pd.DataFrame.from_dict(clf.cv_results_)
results.sort_values(by=['rank_test_score']).head()
```

```python
best_C=clf.best_estimator_.get_params()['C']

best_kernel=clf.best_estimator_.get_params()['kernel']
print('best_C =',best_C,'best_kernel =',best_kernel)
```

```python
from sklearn.svm import SVC
import numpy as np
classifier=SVC(kernel=best_kernel,C=best_C)
classifier.fit(x_train,y_train)

y_predict=classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)

from sklearn.metrics import f1_score
f1_score=f1_score(y_test,y_predict, average=None)
print('f1_score',f1_score)
```

<h2> 4. Dimensional reduction </h2>

```python
#code from : https://www.mikulskibartosz.name/pca-how-to-choose-the-number-of-components/
from sklearn.decomposition import PCA
import numpy as np
pca = PCA().fit(x)
print('pca.explained_variance_ratio_',pca.explained_variance_ratio_)
# % matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,6)

fig, ax = plt.subplots()
xi = np.arange(1, 19, step=1)
yi = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, yi, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 19, step=1)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

```python
pca = PCA(n_components=6)
x_pca=pca.fit_transform(x)
x_pca=pd.DataFrame(x_pca)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.33, random_state=42)
```

<h2>5. Classifier: </h2>

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm, datasets
svc = svm.SVC()
parameters = {'kernel':['linear','rbf'],'C':[1, 2,4,6,10,20,30,50]}
clf = GridSearchCV(svc,parameters)
clf.fit(x_train,y_train)
results = pd.DataFrame.from_dict(clf.cv_results_)
results.sort_values(by=['rank_test_score']).head()
```

```python
best_C=clf.best_estimator_.get_params()['C']
print('best_C =',best_C,'best_kernel =',best_kernel)
best_kernel=clf.best_estimator_.get_params()['kernel']
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel=best_kernel,C=best_C)
classifier.fit(x_train,y_train)

y_predict=classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)

from sklearn.metrics import f1_score
f1_score=f1_score(y_test,y_predict, average=None)
print('f1_score',f1_score)
```

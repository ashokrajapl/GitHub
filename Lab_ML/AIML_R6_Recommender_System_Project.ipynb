<h2> Merge the provided CSVs into one data-frame </h2>

```python
data_frames=[review_1,review_2,review_3,review_4,review_5,review_6]
review = pd.concat(data_frames)
print(review.shape)
```

```python
#impute missing values for scores
import numpy as np
missing = review['score'].isnull()
review.loc[missing,'score']= review.loc[missing,'score'].apply(lambda x: 5.77)
# print(np.sort(review['score'].unique()))

#Round off scores to the nearest integers
review['score']=review['score'].apply(lambda x : round(x))
# print(np.sort(review['score'].unique()))

#Check for duplicate values and remove them if there is any
duplicate = review[review.duplicated()]
duplicate

import re
review=review[['author','product','score']]
review=review.dropna()
review['author']=review['author'].apply(lambda x :re.sub("[^A-za-z0-9\s]","",x))
review['product']=review['product'].apply(lambda x :re.sub("[^A-za-z0-9\s]","",x))

review=review.drop_duplicates(subset=['author','product','score'],keep='first')
duplicate = review[review.duplicated()]
print('Duplicate records',duplicate)
```

<h2> In what business scenario you should use popularity based Recommendation Systems ?</h2>

- Popularity based recommendation systems are used when the business doesnot have any information regarding the characteristics of the user.
- In this case the model reccomends any user the most popular product

<h2>In what business scenario you should use CF based Recommendation Systems ?</h2>

- Collaborative filtering model is used when the business has characteristic details about both the user and the product.so that it can recommend users product based on their similarities with other users.

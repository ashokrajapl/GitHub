```python
laliga_df=pd.read_csv('Laliga.csv',header=1)
#print(laliga)
```

```python
###replace dashes with 0 and change data type to make sure we can perform arithmetic operations on the data.
def change_data_type(column,data):
    for i in data:
        if i =='-':
            laliga_df[column]=laliga_df[column].replace('-',0)
            laliga_df[column]=laliga_df[column].astype(float)            
            break
col_list=list(laliga_df.columns)
print(col_list)
for col in col_list:
    change_data_type(col,list(laliga_df[col]))
    
```

```python
###teams which have started playing between 1930-1980. Use “Debut” column. (Include both the years - 1930 and 1980
###creating a new column Debut_year
deb_yr=[]
def debut_fn(column_name,data):
    for i in data:
        deb_yr.append(i[0:4])
    laliga_df[column_name]=deb_yr
    laliga_df[column_name]=laliga_df[column_name].astype(int)
    
debut_fn('Debut_year',list(laliga_df['Debut']))

###teams which have started playing between 1930-1980

Team_df=laliga_df[(laliga_df.Debut_year >= 1930) & (laliga_df.Debut_year <= 1980)]
Team_df[:][:]
```

```python
###the list of teams which came Top 5 in terms of points 
laliga_df.sort_values('Points',ascending=False).head()
```

```python
###return all the teams with their Goal Differences. Using the same function

def Goal_diff_count(g_for,g_against,team_list):
    g_diff=[]
    team_l=[]
    maxim=0
    minim=0

    for i in range (0,len(g_for)):
        team_l.append(team_list[i])
        g_dif=g_for[i]-g_against[i]
        g_diff.append(g_dif)
    laliga_df['Goal_Difference']=g_diff
    for j in range (0,len(team_l)):
        if g_diff[j]==max(g_diff):
            maxim=j
        elif g_diff[j]==min(g_diff):
            minim=j
    return(g_diff,team_l,maxim,minim)
   
    
    
team_list=list(laliga_df['Team'])       
g_diff,team_l,maxim,minim=Goal_diff_count(list(laliga_df['GoalsFor']),list(laliga_df['GoalsAgainst']),team_list)
print('Team with maximum goal_difference is ',team_l[maxim],'with a difference of ',g_diff[maxim])
print('Team with minimum goal_difference is ',team_l[minim],'with a difference of ',g_diff[minim])
laliga_df.head()


    
```
